```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("violations.csv")

# Plot
plt.figure(figsize=(10,6))
plt.scatter(df["Timestamp_sec"], df["Speed_km_h"])

plt.xlabel("Time (seconds)")
plt.ylabel("Speed (km/h)")
plt.title("Overspeed Violations Over Time")

plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df["Speed_km_h"], bins=10)

plt.xlabel("Speed (km/h)")
plt.ylabel("Number of Vehicles")
plt.title("Distribution of Overspeed Violations")

plt.grid()
plt.show()
```

```python
df["Time_bin"] = df["Timestamp_sec"].astype(int)

count_per_time = df.groupby("Time_bin").size()

count_per_time.plot(kind="bar", figsize=(10,6))

plt.xlabel("Time (seconds)")
plt.ylabel("Number of Violations")
plt.title("Violations Per Time Window")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("violations.csv")

# Style
sns.set(style="whitegrid")

plt.figure(figsize=(12,6))

sns.scatterplot(
    x="Timestamp_sec",
    y="Speed_km_h",
    data=df,
    color="red",
    s=100
)

plt.title("Overspeed Violations Over Time", fontsize=16, fontweight='bold')
plt.xlabel("Time (seconds)", fontsize=12)
plt.ylabel("Speed (km/h)", fontsize=12)

plt.tight_layout()
plt.savefig("output/graphs/speed_vs_time.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.histplot(
    df["Speed_km_h"],
    bins=10,
    kde=True,
    color="orange"
)

plt.title("Distribution of Overspeed Speeds", fontsize=16, fontweight='bold')
plt.xlabel("Speed (km/h)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

plt.tight_layout()
plt.savefig("output/graphs/speed_distribution.png", dpi=300)
plt.show()
```

```python
df["Time_bin"] = df["Timestamp_sec"].astype(int)

count_per_time = df.groupby("Time_bin").size().reset_index(name="Violations")

plt.figure(figsize=(12,6))

sns.barplot(
    x="Time_bin",
    y="Violations",
    data=count_per_time,
    color="blue"
)

plt.title("Violations Per Time Window", fontsize=16, fontweight='bold')
plt.xlabel("Time (seconds)", fontsize=12)
plt.ylabel("Number of Violations", fontsize=12)

plt.tight_layout()
plt.savefig("output/graphs/violations_time.png", dpi=300)
plt.show()
```
